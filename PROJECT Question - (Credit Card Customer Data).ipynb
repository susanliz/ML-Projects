## <span style = "color:orange"> Visualization </span>

### Plot a scatterplot of Avg_credit_Limit vs Total_Credit_Cards with respect to Total_calls_made

```python
plt.figure(figsize=(7,7))
sns.scatterplot(x=df['Avg_Credit_Limit'],y=df['Total_Credit_Cards'],hue=df['Total_calls_made'])
```

### Plot correlation matix on Heatmap

```python
sns.heatmap(df.corr(),annot=True)
```

```python
sns.pairplot(df)
```

### Create K Means model (Let k = 4)

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=4)
```

### Fit predict for our Dataset

```python
cluster = model.fit_predict(df)

cluster
```

### Assign the predicted clusters to a new column in our dataset

```python
df['Cluster'] = cluster
df.head()
```

```python
df.head()
```

## Visualize the clusters

```python
df['Cluster'].unique()
```

```python
# Use x = 'Avg_Credit_Limit' ang y =  'Total_Credit_Cards'

df0 = df[df['Cluster']==0] 
df1 = df[df['Cluster']==1] 
df2 = df[df['Cluster']==2] 
df3 = df[df['Cluster']==3] 
```

```python
df0.columns
```

```python
sns.scatterplot(data=df0,x='Avg_Credit_Limit',y='Total_Credit_Cards',color='green')
sns.scatterplot(data=df1,x='Avg_Credit_Limit',y='Total_Credit_Cards',color='cyan')
sns.scatterplot(data=df2,x='Avg_Credit_Limit',y='Total_Credit_Cards')
sns.scatterplot(data=df3,x='Avg_Credit_Limit',y='Total_Credit_Cards')
```
